```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import median_filter
```

```python
with open ("filtro_1.pkl", mode ='rb') as f:
        series = pickle.load(f) # dados brutos depois do filtro média móvel
```

```python
with open ("filtro_2.pkl", mode ='rb') as f:
        csi = pickle.load(f) #dados brutos sem nenhum fitro
```

```python
def plot(series, title):
	f, ax = plt.subplots()
	plt.plot(series)
	ax.set(xlabel='Samples', ylabel='dB', title=title)
	plt.show()
```

```python
csi = np.array(csi)
csi = csi[0,:,:]
```

```python
series = np.array(series)
series = series[0,:,:]
```

```python
def mediana_filter(series, size):
    return median_filter(series, size=size)
```

## Com filtro média móvel

```python
plot(series, "Filtro Media Móvel")
```

```python
plot(mediana_filter(series, 10), "Filtro mediana")
```

## Sem filtro

```python
plot(csi, "Sem filtro")
```
